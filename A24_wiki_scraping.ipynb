# Scraping A24 Filmography Wikipedia Page + Pandas

```python
# Load libraries

from bs4 import BeautifulSoup
import requests
import pandas as pd
```

```python
# url address 
url = 'https://en.wikipedia.org/wiki/List_of_A24_films'
# request information from url
web_page = requests.get(url)
# convert to BS object
soup = BeautifulSoup(web_page.text, 'html.parser')
```

```python
print(soup.prettify())
```

```python
soup.find('title')
```

```python
print(soup.title.text)
```

```python
soup.find_all('table')
```

# Films Released in 2010s

```python
table2010 = soup.find_all('table')[0]
print(table2010)
```

```python
table2010.find('tr')
```

```python
col_titles2010 = table2010.find('tr')
col_titles2010.find_all('th')
```

```python
col_table2010_titles = [title.text.strip() for title in col_titles2010]

print(col_table2010_titles)
```

```python
df2010 = pd.DataFrame(columns = col_table2010_titles)

df2010.drop(['Ref.', ''], axis = 1)
```

```python
table2010.find_all('tr')
```

```python
col_data2010 = table2010.find_all('tr')
```

```python
df2010 = pd.read_html(url)[0].reset_index()
df2010.columns.header = df2010.loc[0]
df2010 = df2010[0:]
df2010
```

```python
df2010.to_csv(r'C:\Users\Janis\OneDrive\Documents\Data Analytics Projects\Portfolio Case Studies\Case Study-A24\A24_Films_2010s.csv', index=False)
```
